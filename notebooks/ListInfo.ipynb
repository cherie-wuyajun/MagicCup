```python
import pandas as pd
import numpy as np
%matplotlib inline
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
listing_info_df = pd.read_csv('dataset/listing_info.csv',parse_dates=['auditing_date'])
listing_info_df.head()
```

```python
corr=listing_info_df.corr()
# plot the heatmap
sns.heatmap(corr, 
        xticklabels=corr.columns,
        yticklabels=corr.columns)
```

```python
list_term=listing_info_df['term'].value_counts()
list_term
```

```python
list_principle=listing_info_df['principal'].value_counts()
list_principle
```

```python
principal_df=pd.DataFrame(list_principle)
principal_df.head()
```

```python
principal_df.rename(columns={principal_df.columns[0]: "frequency"},inplace=True)
principal_df['principal']=principal_df.index
principal_df
```

```python
principal_df.reset_index(drop=True, inplace=True)
principal_df
```

```python
label=principal_df.principal
size=principal_df.frequency
plt.figure(figsize=(10,10))
plt.pie(size,labels=label,autopct='%1.11f%%')
plt.axis('equal')
plt.show()
```

```python
def ChangeToClass(principal):
    if principal<=1170:
        return 6
    elif 1170<principal<=2260:
        return 5
    elif 2260<principal<=3350:
        return 4
    elif 3350<principal<=3910:
        return 3
    elif 3910<principal<=5540:
        return 2
    else:
        return 1
```

```python
principal_df['principal_class']=principal_df['principal'].apply(ChangeToClass)
del principal_df['frequency']
principal_df
```

```python
listing_info_df=listing_info_df.merge(principal_df, on="principal", how="left")
listing_info_df
```

```python
listing_info_df['TermRate']=listing_info_df['rate']/listing_info_df['term']
listing_info_df
```

```python
listing_info_df['TermRate'].value_counts()
```

```python
corr1=listing_info_df.corr()
# plot the heatmap
sns.heatmap(corr1, 
        xticklabels=corr1.columns,
        yticklabels=corr1.columns)
```
