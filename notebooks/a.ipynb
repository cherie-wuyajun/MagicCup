```python
bin_size=(1,2,4,8)
def count_to_style(count):
    n=1
    num=1
    while count>n and num<5:
        n=n*2
        num=num+1
    return num
```

```python
user_unique['loan_count']=user_unique['count'].apply(count_to_style)
```
